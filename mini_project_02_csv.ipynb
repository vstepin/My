```python
df.columns.tolist()
```

```python
df.rename(
    columns = {
      'Дата продажу'  : 'date_of_sale',
      'Назва курсу'   : 'course',
      'Вартість$'     : 'cost',
      'Кіл-ть'        : 'quantity',
      'Всього'        : 'total'
    },
    inplace = True
)

df.head(10)
```

```python
df.info()
```

```python
df.date_of_sale.unique()
```

```python
df['date_of_sale'] = df.date_of_sale.apply(lambda item: parse(item).strftime('%Y-%m-%d'))
df['date_of_sale'] = pd.to_datetime(df.date_of_sale)
```

```python
df.info()
```

```python
df.course.unique()
```

```python
df.course.replace(
  {
      'Python нач'              : 'python for beginners',
      'DS Boot'                 : 'data science bootcamp',
      'Web-разработка'          : 'web development',
      'Machine Learning ADV'    : 'machine learning advanced',
      'SQL БАЗОВЫЙ'             : 'sql crash course',
      'Python For Beginners'    : 'python for beginners',
      'Быстрый курс SQL'        : 'sql crash course',
      'Data Science BOOTCAMP'   : 'data science bootcamp',
      'Web Dev 101'             : 'web development',
      'Продвинутый ML'          : 'machine learning advanced',
      'Питон для начинающих'    : 'python for beginners',
      'Данные.Наука'            : 'data science bootcamp',
      'sql-cc'                  : 'sql crash course',
      'веб разработка'          : 'web development',
      'SQL-Курс'                : 'sql crash course',
      'WEB DEVELOPMENT'         : 'web development',
      'ML (уровень 2)'          : 'machine learning advanced',
      'web_dev'                 : 'web development',
      'машинное обучение+'      : 'machine learning advanced',
      'Буткемп по Data Science' : 'data science bootcamp',
      'PYTHON-BEGINNERS'        : 'python for beginners'
  },
  inplace = True
)

df.course.unique()
```

```python
df['course'] = df.course.astype('category')
df.info()
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(inplace = True)
df.info()
```

```python
df.describe()
```

```python
df.loc[df.isna().any(axis=1)]
```

```python
df['quantity'] = df.quantity.fillna(df.total / df.cost)
```

```python
df['quantity'] = df.quantity.astype('int64')
```

```python
df[df.quantity == 0]
```

```python
df = df[df.quantity > 0]
df.info()
```

```python
df.describe()
```

```python
df.groupby('course').quantity.mean()
```
