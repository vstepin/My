```python
df.rename(
    columns = {
      'Дата продажу': 'date_of_sale',
      'Назва курсу': 'course',
      'Вартість$': 'cost',
      'Кіл-ть': 'quantity',
      'Всього': 'total'
    },
    inplace = True
)

df.head(10)
```

```python
df.info()
```

```python
df.date_of_sale.unique()
```

```python
df['date_of_sale'] = df.date_of_sale.apply(lambda item: parse(item).strftime('%Y-%m-%d'))
df['date_of_sale'] = pd.to_datetime(df.date_of_sale)
```
